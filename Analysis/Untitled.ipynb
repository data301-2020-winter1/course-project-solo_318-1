```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from Scripts import poject # This is called a relative import
df = poject.load_and_process(r'\Users\john\Desktop\course-project-solo_318-1\Analysis\listings(2).csv')

sns.set_theme(style="ticks",
              font_scale=1.3,
             )

plt.rc("axes.spines", top=False, right=False)
ax = sns.countplot(y='room_type', data=df, order=["Entire home/apartment", "Private room", "Shared room", "Hotel room"])
ax.set_xlim(0,12000)
plt.title('Room types and their prices')
ax.set(xlabel='Price($)', ylabel='Room type')
plt.show()
```
